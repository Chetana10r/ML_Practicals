Dataset1:  Employee Salary dataset

```python
#1.	Display structure and summary report.

import pandas as pd

data = pd.read_csv("Employee_Salary_Dataset.csv")
print(data)
print(data.info())
print(data.describe())  
```

```python
#2.	Check data types of each column and ensure whether the data types for each of the attributes are appropriate, if not make appropriate corrections.

print(data.dtypes)
```

```python
#3.	Display NULL values for each of the columns

print(data.isnull().sum())
```

```python
#4.	Find out missing values and replace them for the applicable columns.
    
data.columns = data.columns.str.strip()

data['Salary'] = data['Salary'].fillna(data['Salary'].mean())

print(data)
```

```python
# 5. Find out categorical data and perform encoding on relevant features.

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

data['Gender'] = label_encoder.fit_transform(data['Gender'])

print("\nData after encoding 'Gender':")
print(data)
```

```python
print("After removing null values: \n")

print(data.isnull().sum())
```

#Dataset2:  Titanic  Dataset

```python
#1.	Display structure and summary report.

import pandas as pd

data1 = pd.read_csv("Titanic_Dataset.csv")
print(data1)
print(data1.info())
print(data1.describe()) 
```

```python
#2.	Check data types of each column and ensure whether the data types for each of the attributes are appropriate, if not make appropriate corrections.

print(data1.dtypes)
```

```python
#3.	Display NULL values for each of the columns

print(data1.isnull().sum())

print(data1.columns)
```

```python
#4.	Find out missing values and replace them for the applicable columns.
    
data1.columns = data1.columns.str.strip()

if data1['Age'].dtype != 'float64': 
    data1['Age'] = data1['Age'].astype('float64')

data1['Age'] = data1['Age'].fillna(data1['Age'].mean())

data1['Fare'] = data1['Fare'].fillna(data1['Fare'].mean())


print(data1)
```

```python
# 5. Find out categorical data and perform encoding on relevant features.

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

data1['Cabin'] = label_encoder.fit_transform(data1['Cabin'])

print("\nData after encoding 'Cabin':")
print(data1)
```

```python
print("After removing null values: \n")

print(data1.isnull().sum())
```

#Dataset3:  NYC AirBnB Dataset

```python
#1.	Display structure and summary report.

import pandas as pd

data2 = pd.read_csv("NYC AirBnB Dataset.csv")
print(data2)
print(data2.info())
print(data2.describe()) 
```

```python
#2.	Check data types of each column and ensure whether the data types for each of the attributes are appropriate, if not make appropriate corrections.

print(data2.dtypes)
```

```python
#3.	Display NULL values for each of the columns

print(data2.isnull().sum())
```

```python
#4.	Find out missing values and replace them for the applicable columns.
    
data2.columns = data2.columns.str.strip()

data2['reviews_per_month'] = data2['reviews_per_month'].fillna(data2['reviews_per_month'].mean())


print(data2)
```

```python
# 5. Find out categorical data and perform encoding on relevant features.

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

data2['last_review'] = label_encoder.fit_transform(data2['last_review'])

data2['host_name'] = label_encoder.fit_transform(data2['host_name'])

data2['name'] = label_encoder.fit_transform(data2['name'])

print("\nData after encoding 'Last_Review':")

print(data2)
```

```python
print("After removing null values: \n")

print(data2.isnull().sum())
```
