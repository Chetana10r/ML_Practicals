Dataset1:  Employee Salary dataset

```python
#1.	Display structure and summary report.

import numpy as np
import pandas as pd

data = pd.read_csv("Employee_Salary_Dataset.csv")
print(data)
print(data.info())
print(data.describe())  
```

```python
#2.	Check data types of each column and ensure whether the data types for each of the attributes are appropriate, if not make appropriate corrections.

print(data.dtypes)
```

```python
#Function to validate text for gender

def validate_gender(value):
    valid_genders= {"male","female","other"}
    return value if isinstance(value, str) and value.lower() in valid_genders else "Unknown"

#Function to validate text for ID

def validate_int(value):
    return value if isinstance(value, int) else np.nan

#Function to validate float values

def validate_float(value):
    return value if isinstance(value, float) else np.nan

#Function to validate text fields

def validate_text(value):
    return value if isinstance(value, str) else np.nan
```

```python
#Applying the validation rules

data['ID']= data['ID'].apply(validate_int)
data['Experience_Years']= data['Experience_Years'].apply(validate_int)
data['Age']= data['Age'].apply(validate_float)
data['Gender']= data['Gender'].apply(validate_gender)
data['Salary']= data['Salary'].apply(validate_int)
```

```python
#3.	Display NULL values for each of the columns

print(data.isnull().sum())
```

```python
#4.	Find out missing values and replace them for the applicable columns.
    
data.columns = data.columns.str.strip()

data['Salary'] = data['Salary'].fillna(data['Salary'].mean())

print(data)
```

```python
# 5. Find out categorical data and perform encoding on relevant features.

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

data['Gender'] = label_encoder.fit_transform(data['Gender'])

print("\nData after encoding 'Gender':")
print(data)
```

```python
print("After removing null values: \n")

print(data.isnull().sum())
```
