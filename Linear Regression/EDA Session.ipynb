```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from Seaborn
titanic = sns.load_dataset("titanic")

# Example 1: Univariate analysis on 'age' (numerical variable)
plt.figure(figsize=(8, 5))
sns.histplot(titanic['age'],color='blue')
plt.title('Distribution of Age on Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

"""Most passengers were between 20 and 40 years old, as this is where the highest bars are"""


"""This means the Titanic mostly carried working-age adults, with some families (young children) and older individuals"""


# Example 2: Univariate analysis on 'pclass' (categorical variable)
plt.figure(figsize=(8, 5))
sns.countplot(x='pclass', data=titanic)
plt.title('Distribution of Passenger Class')
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Count')
plt.show()



"""First Class (Pclass 1) had fewer passengers, indicating it was more exclusive and expensive"""



# Example 3: Univariate analysis on 'fare' (numerical variable)
plt.figure(figsize=(8, 5))
sns.histplot(titanic['fare'])
plt.title('Distribution of Fare on Titanic')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

"""This shows a big gap between the majority of budget travelers and a small number of wealthy passengers"""

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset("titanic")



# Example: Comparison of Age by Survival

sns.kdeplot(data=titanic, x="age", hue="survived", common_norm=False,shade=True)  # shade=True
plt.title('Density Plot of Age by Survival Status')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend(title="Survived", labels=["Did Not Survive", "Survived"])
plt.show()


"""Insight: Younger passengers were more likely to survive, possibly because they received priority during evacuation (e.g., "women and children first")."""
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Example 1: Barplot for total passengers by embarkation point
plt.figure(figsize=(8, 5))
sns.countplot(data=titanic, x="embarked", palette="viridis")
plt.title("Number of Passengers by Embarkation Point", fontsize=14)
plt.xlabel("Embarkation Point", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Example 2: Barplot for survival by embarkation point
plt.figure(figsize=(8, 5))
sns.barplot(data=titanic, x="embarked", y="survived", ci=None, palette="Set2")
plt.title("Survival Rate by Embarkation Point", fontsize=14)
plt.xlabel("Embarkation Point", fontsize=12)
plt.ylabel("Survival Rate", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Load Titanic dataset
from seaborn import load_dataset
titanic = load_dataset("titanic")

# Count the number of passengers in each class
class_counts = titanic['class'].value_counts()

# Pie chart for Passenger Class distribution
plt.pie(
    class_counts, 
    labels=class_counts.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['gold', 'skyblue', 'lightgreen'],
    explode=(0.05, 0, 0)  # Slightly separate the First Class for emphasis
)
plt.title("Passenger Class Distribution on the Titanic")
plt.show()
```

# Bivariate Analysis

```python
#Bivariate analysis depends on the type of data and it uses two variables
#scatterplot: reln between two variables
#line chart: trend vs time
#Heatmap: relationship between numeric values
#Barplot(grouped): Compare distributions(surv vc not surviived)
#violin plot: combine compared distribution
#Pair plot: multiple plots
#Multivariate Analysis: more than 2 bivariate analysis(2 variables)
```

```python
import matplotlib.pyplot as plt
from seaborn import load_dataset

# Load Titanic dataset
titanic = load_dataset("titanic")

# Scatter plot for Age vs Fare
plt.figure(figsize=(10, 6))
plt.scatter(titanic['age'], titanic['fare'], alpha=0.6, color='purple')

# Adding title and labels
plt.title("Scatter Plot of Age vs Fare on the Titanic", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Fare", fontsize=12)

# Display plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset

# Load Titanic dataset
titanic = load_dataset("titanic")

# Group by 'class' and calculate the average 'fare'
class_fare = titanic.groupby('class')['fare'].mean().reset_index()

# Line chart for Average Fare by Class

sns.lineplot(data=class_fare, x='class', y='fare')

# Adding title and labels
plt.title("Average Fare by Passenger Class on the Titanic")
plt.xlabel("Class")
plt.ylabel("Average Fare")

# Display plot
plt.show()


"""This line chart effectively summarizes the relationship between passenger class and fare, showing a neat and intuitive comparison across categories"""
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset

# Load Titanic dataset
titanic = load_dataset("titanic")

# Select numerical columns for correlation
numerical_data = titanic[['age', 'fare', 'sibsp', 'parch', 'survived']]

# Calculate correlation matrix
corr_matrix = numerical_data.corr()

# Plotting heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Adding title
plt.title("Correlation Heatmap of Titanic Dataset")

# Display plot
plt.show()


#Insights:

#There is likely no strong correlation between Age and Fare
#Survival and Fare may show a weak to moderate positive correlation
#Survival and Age might show a weak negative or near-zero correlation.
#The correlation between SibSp and Parch is expected to be positive
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset

# Load Titanic dataset
titanic = load_dataset("titanic")

# Boxplot for Fare vs Survival
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='survived', y='fare', palette='Set2')

# Adding title and labels
plt.title("Boxplot of Fare vs Survival on the Titanic")
plt.xlabel("Survival (1 = Survived, 0 = Did not survive)")
plt.ylabel("Fare")

# Display plot
plt.show()


#Higher-paying passengers (likely in higher classes) were more likely to survive, as indicated by the higher median fare and wider spread for survivors.

#Non-survivors generally paid lower fares, suggesting they were more likely to be in lower classes.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset

# Load Titanic dataset
titanic = load_dataset("titanic")

# Violin plot for Fare vs Survival
plt.figure(figsize=(10, 6))
sns.violinplot(data=titanic, x='survived', y='fare', palette='Set2')

# Adding title and labels
plt.title("Violin Plot of Fare vs Survival on the Titanic", fontsize=16)
plt.xlabel("Survival (1 = Survived, 0 = Did not survive)", fontsize=12)
plt.ylabel("Fare", fontsize=12)

# Display plot
plt.show()

"""The violin plot reinforces the insight that passengers who paid higher fares (likely from higher classes like First Class) were more likely to survive."""

"""It also clearly shows that the majority of non-surviving passengers paid lower fares, suggesting they were likely in Third Class or the lower part of Second Class."""
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset

# Load Titanic dataset
titanic = load_dataset("titanic")

# Select relevant columns for pairplot
pair_data = titanic[['fare', 'age', 'survived']]

# Pair plot for multivariate analysis
sns.pairplot(pair_data, hue='survived')

# Display the plot
plt.show()


#Higher fares tend to correlate with higher chances of survival, especially in First Class.
#Younger passengers (especially children) had a higher survival rate, although the fare still played a significant role.
#The pair plot gives a clear visual of how Age and Fare interact with Survival.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from seaborn import load_dataset

# Load Titanic dataset
titanic = load_dataset("titanic")

# Select numerical columns for multivariate analysis
numerical_data = titanic[['age', 'fare', 'sibsp', 'parch', 'survived']].dropna()

# Calculate the correlation matrix
corr_matrix = numerical_data.corr()

# Plotting the heatmap

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Adding title
plt.title("Correlation Heatmap of Titanic Dataset (Numerical Features)", fontsize=16)

# Display plot
plt.show()

# Display coefficient matrix
print("Coefficient Matrix (Correlation Matrix):")
print(corr_matrix)


"""The strongest positive correlation is between Fare and Survival, indicating that higher fares are associated with higher survival rates"""
# Age has a weak negative correlation with survival, suggesting that younger passengers were more likely to survive.
```
